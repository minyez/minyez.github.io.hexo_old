```python
'''Meshgrid plot
'''
import matplotlib.pyplot as plt
import numpy as np

nx = 100
ny = 100

u = np.linspace(-2, 2, nx)
v = np.linspace(-2, 2, ny)

x, y = np.meshgrid(u, v)
```

```python
z = (x+y)*np.exp(-x**2-y**2)
```

```python
plt.set_cmap('gray')
plt.pcolor(z)
plt.colorbar()
plt.show()
```
